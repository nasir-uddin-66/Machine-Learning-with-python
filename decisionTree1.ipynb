# **Decision Tree**

## Import Necessary Libraries

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
import matplotlib.pyplot as plt
```

## Read csv file and create Dataframe

```python
df = pd.read_csv('shop data.csv')
df.head()
```

## Making pie chart

```python
counts = df['buys'].value_counts()
counts.plot.pie(autopct='%1.1f%%', figsize=(5, 5), startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('Yes/No Responses')
plt.ylabel('')
plt.show()
```

## Encoding Values

```python
# label encoding
labelEncoder = LabelEncoder()
df['age'] = labelEncoder.fit_transform(df['age'])
df['gender'] = labelEncoder.fit_transform(df['gender'])
df['m_status'] = labelEncoder.fit_transform(df['m_status'])
df['buys'] = labelEncoder.fit_transform(df['buys'])
df['income'] = labelEncoder.fit_transform(df['income'])
```

```python
df.head()
```

## Train-Test splitting data

```python
xtrain,xtest,ytrain,ytest = train_test_split(df[['age', 'income', 'gender', 'm_status']],df['buys'],test_size=.25,random_state=1)
```

## Training model

```python
classifier = DecisionTreeClassifier()
classifier.fit(xtrain,ytrain)
```

## Testing

```python
p = classifier.predict(xtest)
print(p)
```

## Score

```python
classifier.score(xtest,ytest)
```
